```python
import numpy as np
import pandas as pd
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient(alpha,start,max_iter):
    x_list=list()
    x=start
    x_list.append(x)
    for i in range(max_iter):
        gradi=derivative(x)
        x=x-(alpha*gradi)
        x_list.append(x)
    return x_list
x=sym.symbols('x')
expr=(x+3)**2.0
grad=sym.Derivative(expr,x)
print("{}".format(grad.doit()))
grad.doit().subs(x,2)
```

```python
alpha=0.1
start=2
max_iter=30
x=sym.symbols('x')
expr=(x+3)**2
```

```python
x_cor=np.linspace(-15,15,100)
plt.plot(x_cor,objective(x_cor))
plt.plot(2,objective(2),'ro')
```

```python
x=gradient(alpha,start,max_iter)
x_cor=np.linspace(-5,5,100)
plt.plot(x_cor,objective(x_cor))
```

```python
x_arr=np.array(x)
plt.plot(x_arr,objective(x_arr),'.-',color='red')
plt.show()
```
